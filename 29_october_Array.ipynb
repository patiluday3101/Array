```python
# #1. Given an array, check if it contains any duplicates or not.
# arr = [1, 2, 4, 2, 5, 9]
# Output = True
```

```python
def find_duplicate(arr):
  seen=set() # Initialize an empty set to store seen elements
  for i in arr: # Iterate through the array
    if i in seen: # Check if the current element has already been seen
      return True # If seen, return True indicating a duplicate exists
    seen.add(i) # If not seen, add the element to the set
  return False # If no duplicates were found, return False



arr= [1,2,4,2,5,9] # Example array
find_duplicate(arr)

if find_duplicate(arr):
  print("True") # Print True if duplicates exist
else:
  print("False") # Print False if no duplicates exist
```

```python
# 2. Given an array and an integer k, rotate the array to the right by k steps.
# arr = [1, 2, 3, 4, 5, 6, 7] k = 3
# Output = [5, 6, 7, 1, 2, 3, 4]
```

```python
def rotate(arr,k):
  n=len(arr) # Get the length of the array
  rotated_arr=[None]*n # Initialize a new array of the same length
  for i in range(n): # Iterate over each element in the original array
    new_index =(i+k)%n # Calculate the new index after rotation
    rotated_arr[new_index]=arr[i] # Place the element at its new index
  return rotated_arr # Return the rotated array


arr = [1, 2, 3, 4, 5, 6, 7] # Example array
k = 3 # Number of steps to rotate
rotated_arr = rotate(arr, k)
print(rotated_arr) # Print the rotated array. Execute this code to see the output.
```

```python
# 3. Reverse the given array in-place, means without using any extra data structure.
# arr = [2, 4, 5, 7, 9, 12]
# Output = [12, 9, 7, 5, 4, 2]
```

```python
def reverse(arr):
  n=len(arr) # Get the length of the array
  for i in range(n//2): # Iterate over half of the array
    arr[i],arr[n-i-1]=arr[n-i-1],arr[i] # Swap the elements at symmetrical positions
  return arr # Return the reversed array
arr = [2, 4, 5, 7, 9, 12]
reverse(arr) # Execute this code to see the output.
```

```python
# 4. Given an array of integers, find the maximum element in an array
# arr = [10, 5, 20, 8, 15]
# Output = 20
```

```python
def find_max(arr):
  n=len(arr) # Get the length of the array
  max_element=arr[0] # Initialize the maximum element with the first element
  for i in range(1,n): # Iterate over the remaining elements in the array
    if arr[i]>arr[0]: # Check if the current element is greater than the first element
      max_element=arr[i] # Update the first element with the current element
      return max_element # Return the maximum element
arr= [10, 5, 20, 8, 15]
find_max(arr) # Execute this code to see the output.
```
